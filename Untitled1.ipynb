```python
df['request_time'] = pd.to_datetime(df['request_time'])
df['ahm_pickup_time'] = pd.to_datetime(df['ahm_pickup_time'])
df['create_time'] = pd.to_datetime(df['create_time'])

# biến đổi dữ liệu về timestamp
```

```python
df['request_date'] = df['request_time'].dt.date
df['ahm_pickup_date'] = df['ahm_pickup_time'].dt.date
df['create_date'] = df['create_time'].dt.date

# trích xuất ngày tháng nắm từ cột
```

```python
df
```

```python
df['diff_date'] = df['create_date'] - df['ahm_pickup_date']
df['day'] = df['diff_date'].dt.days
df

# tạo ra cột mới để xem có bao nhiêu đơn có pickup và create cùng ngày, bao nhiêu đơn là N+1, N+2
```

```python
df['request_time'] = df['request_time'].dt.time
df['ahm_pickup_time'] = df['ahm_pickup_time'].dt.time
df['create_time'] = df['create_time'].dt.time
```

```python
df
```

```python
df_pickup = df.groupby(['subject','ahm_pickup_date'])['tracking_number'].count().reset_index(name = 'num_tracking_pickupdate')
df = pd.merge(df, df_pickup, on = ['subject','ahm_pickup_date'], how = 'left')
df
```

```python
df_day = df.groupby(['subject','day'])['tracking_number'].count().reset_index(name = 'num_tracking_diff_day')
df_day

# thống kê đơn theo subject và diff_day

# vì hiện nay số lượng đơn pickup vào N và create từ N+4 trở đi khá ít 
# nên khi chia nhỏ thành các tập thì nên tập trung láy từ ngày N đến N+1 và các ngày sau đó sẽ gộp lại thành allN
```

```python
df_gr = df.groupby(['subject','day'])['tracking_number'].count().reset_index(name = 'num_tracking_diff_day')
df = pd.merge(df, df_gr, on = ['subject','day'], how = 'left')
df
```

```python
df_sub = df.groupby('subject')['tracking_number'].count().reset_index(name = 'num_tracking_sub')
df_sub

# thống kê số lượng đơn theo Normal và Reseller
```

```python
df.info()
```

```python
df
```

```python
cutoff_01 = '11:00:00'
cutoff_02 = '14:00:00'
cutoff_03 = '17:00:00'
cutoff_04 = '20:00:00'

# tạo ra 4 mốc thời gian giống thời gian lên đơn của lazada để lượng hóa giờ pickup và create đơn về 4 khung này
```

```python
df['ahm_pickup_time'] = df['ahm_pickup_time'].astype('str')
df['create_time'] = df['create_time'].astype('str')

# ép kiểu để có thể lượng hóa 
```

```python
df
```

```python
col_convert = ['ahm_pickup_time', 'create_time']
for col in col_convert:
    for i in range(len(df)):
        if df[col][i] <= cutoff_01:
            df[col][i] = '11:00:00'
        elif df[col][i] <= cutoff_02:
            df[col][i] = '14:00:00'
        elif df[col][i] <= cutoff_03:
            df[col][i] = '17:00:00'
        elif df[col][i] <= cutoff_04:
            df[col][i] = '20:00:00'
        elif df[col][i] > cutoff_04:
            df[col][i] = '20:00:00'
            
# biến đổi thời gian phân bổ về 4 khung giờ trên 
```

```python
df
```

```python
# chia thành các dataframe nhỏ mỗi tương ứng

# df_01 là đơn có pickup và create cùng ngày N

# df_02 là đơn có pickup ngày N và create N+1

# df_03 là đơn có pickup ngày N và create N+2

# df_04 là đơn có pickup ngày N và create N+3


# theo như thống kê 
```

```python
df_01 = (df[df.day == 0]).reset_index(drop = True)
```

```python
df_02 = (df[df.day == 1]).reset_index(drop = True)
```

```python
df_03_01 = (df[df.day == 2]).reset_index(drop = True)
df_03_02 = (df[df.day == 3]).reset_index(drop = True)
df_03_03 = (df[df.day == 4]).reset_index(drop = True)
df_03_04 = (df[df.day == 5]).reset_index(drop = True)
df_03_05 = (df[df.day == 6]).reset_index(drop = True)
```

```python
df_01
```

```python
df_time_01 = df_01.groupby(['subject','ahm_pickup_time','create_time'])['tracking_number'].count().reset_index(name = 'num_tracking_N')
df_time_01
```

```python
df_01_normal = df_time_01[df_time_01.subject == 'Normal']

df_01_normal['rate_N'] = df_01_normal['num_tracking_N'] / df_sub['num_tracking_sub'][0] * 100
```

```python
df_time_02 = df_02.groupby(['subject','ahm_pickup_time','create_time'])['tracking_number'].count().reset_index(name = 'num_tracking_N+1')
df_time_02
```

```python
df_02_normal = df_time_02[df_time_02.subject == 'Normal']

df_02_normal['rate_N+1'] = df_02_normal['num_tracking_N+1'] / df_sub['num_tracking_sub'][0] * 100
```

```python
df_02_normal
```

```python
df_01_normal
```

```python
df_03 = pd.concat([df_03_01, df_03_02, df_03_03, df_03_04, df_03_05], axis = 0)

df_time_03 = df_03.groupby(['subject','ahm_pickup_time','create_time'])['tracking_number'].count().reset_index(name = 'num_tracking_allN')
df_time_03
```

```python
df_03_normal = df_time_03[df_time_03.subject == 'Normal']

df_03_normal['rate_allN'] = df_03_normal['num_tracking_allN'] / df_sub['num_tracking_sub'][0] * 100
df_03_normal
```

```python
df_merge = pd.merge(df_01_normal, df_02_normal, on = ['ahm_pickup_time', 'create_time'], how = 'outer')
df_merge = pd.merge(df_merge, df_03_normal, on = ['ahm_pickup_time', 'create_time'], how = 'outer')
```

```python
df_merge
```

```python
df_merge = df_merge.sort_values(by = ['ahm_pickup_time','create_time'])
```

```python
df_merge = df_merge.fillna(0)
```

```python
df_merge 
```

```python
col_normal_01 = list(df_merge['rate_N'])

col_normal_01_col_01 = col_normal_01[:4]

col_normal_01_col_02 = col_normal_01[4:8]

col_normal_01_col_03 = col_normal_01[8:12]

col_normal_01_col_04 = col_normal_01[12:]

col_normal_01_col_01
```

```python
col_normal_02 = list(df_merge['rate_N+1'])

col_normal_02_col_01 = col_normal_02[:4]

col_normal_02_col_02 = col_normal_02[4:8]

col_normal_02_col_03 = col_normal_02[8:12]

col_normal_02_col_04 = col_normal_02[12:]

col_normal_02_col_01
```

```python
col_normal_03 = list(df_merge['rate_allN'])

col_normal_03_col_01 = col_normal_03[:4]

col_normal_03_col_02 = col_normal_03[4:8]

col_normal_03_col_03 = col_normal_03[8:12]

col_normal_03_col_04 = col_normal_03[12:]

col_normal_03_col_01
```

```python
col_normal_01 = col_normal_01_col_01 + col_normal_02_col_01 + col_normal_03_col_01
```

```python
col_normal_02 = col_normal_01_col_02 + col_normal_02_col_02 + col_normal_03_col_02
col_normal_03 = col_normal_01_col_03 + col_normal_02_col_03 + col_normal_03_col_03
col_normal_04 = col_normal_01_col_04 + col_normal_02_col_04 + col_normal_03_col_04
```

```python
# col = ['11:00:00','14:00:00','17:00:00','20:00:00','11:00:00+N1','14:00:00+N1','17:00:00+N1','20:00:00+N1',
#       '11:00:00+allN','14:00:00+allN','17:00:00+allN','20:00:00+allN']

col = ['11H','14H','17H','20H','11H + N1','14H + N1','17H + N1','20H + N1',
      '11H + allN','14H + allN','17H + allN','20H + allN']

# row = ['11:00:00','14:00:00','17:00:00','20:00:00']

row = ['11H','14H','17H','20H']
```

```python
param = [col_normal_01, col_normal_02, col_normal_03, col_normal_04]
f_all = pd.DataFrame(param, index = list(row), columns = list(col))
f_all
```

```python
df_01_reseller = df_time_01[df_time_01.subject == 'Reseller']

df_01_reseller['rate_N'] = df_01_reseller['num_tracking_N'] / df_sub['num_tracking_sub'][1] * 100

df_02_reseller = df_time_02[df_time_02.subject == 'Reseller']

df_02_reseller['rate_N+1'] = df_02_reseller['num_tracking_N+1'] / df_sub['num_tracking_sub'][1] * 100

df_03_reseller = df_time_03[df_time_03.subject == 'Reseller']

df_03_reseller['rate_allN'] = df_03_reseller['num_tracking_allN'] / df_sub['num_tracking_sub'][1] * 100
df_03_reseller

df_merge_reseller = pd.merge(df_01_reseller, df_02_reseller, on = ['ahm_pickup_time', 'create_time'], how = 'outer')
df_merge_reseller = pd.merge(df_merge_reseller, df_03_reseller, on = ['ahm_pickup_time', 'create_time'], how = 'outer')

df_merge_reseller = df_merge_reseller.sort_values(by = ['ahm_pickup_time','create_time'])

df_merge_reseller = df_merge_reseller.fillna(0)

col_reseller_01 = list(df_merge_reseller['rate_N'])
col_reseller_01_col_01 = col_reseller_01[:4]
col_reseller_01_col_02 = col_reseller_01[4:8]
col_reseller_01_col_03 = col_reseller_01[8:12]
col_reseller_01_col_04 = col_reseller_01[12:]

col_reseller_02 = list(df_merge_reseller['rate_N+1'])
col_reseller_02_col_01 = col_reseller_02[:4]
col_reseller_02_col_02 = col_reseller_02[4:8]
col_reseller_02_col_03 = col_reseller_02[8:12]
col_reseller_02_col_04 = col_reseller_02[12:]

col_reseller_03 = list(df_merge_reseller['rate_allN'])
col_reseller_03_col_01 = col_reseller_03[:4]
col_reseller_03_col_02 = col_reseller_03[4:8]
col_reseller_03_col_03 = col_reseller_03[8:12]
col_reseller_03_col_04 = col_reseller_03[12:]

col_reseller_01 = col_reseller_01_col_01 + col_reseller_02_col_01 + col_reseller_03_col_01
col_reseller_02 = col_reseller_01_col_02 + col_reseller_02_col_02 + col_reseller_03_col_02
col_reseller_03 = col_reseller_01_col_03 + col_reseller_02_col_03 + col_reseller_03_col_03
col_reseller_04 = col_reseller_01_col_04 + col_reseller_02_col_04 + col_reseller_03_col_04
```

```python
param_reseller = [col_reseller_01, col_reseller_02, col_reseller_03, col_reseller_04]
f_all_reseller = pd.DataFrame(param_reseller, index = list(row), columns = list(col))
f_all_reseller
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (20,10))
img = sns.heatmap(f_all, vmax = 10, center = 0, fmt = '.4f', cmap = 'coolwarm',
           square = True, linewidths = .2, annot = True, cbar_kws = {'shrink': .7})

ax.set_title('Phân Bổ Thời Gian Đơn Hàng Normal (%)', fontsize = 30)
```

- Khối lượng hàng được pickup và tạo đơn phần lớn thường được diễn ra cùng ngày và dàn trải đều vào các khung giờ.

- Đối với các đơn hàng được pickup vào khung giờ 11H thì sẽ được xử lý khá nhanh và create vào khung giờ 11H (7,2%) và đặc biệt là phần lớn vào 14H (12,8%) cùng ngày.

- Đối với hàng pickup trong khung giờ 14H và 17H thì phần lớn sẽ tạo lần lượt vào khung 17H (8,3%) và 20H (14,2%) cùng ngày. Ngoài ra một phần hàng được pickup vào khung giờ 17H và 20H sẽ được create vào 11H (7,1% và 7,4%) ở ngày tiếp theo.

```python
fig, ax = plt.subplots(figsize = (20,10))
img_reseller = sns.heatmap(f_all_reseller, vmax = 5, center = 0, fmt = '.4f', cmap = 'coolwarm',
           square = True, linewidths = .2, annot = True, cbar_kws = {'shrink': .7})

ax.set_title('Phân Bổ Thời Gian Đơn Hàng Reseller (%)', fontsize = 30)
```

- Trong tháng 6 số lượng đơn Reseller cao gấp 3,5 lần số lượng hàng Normal. Vì bản chất hàng Reseller là mặt hàng chiếm số lượng lớn. Nên thời gian phân bổ của hàng Reseller cũng khác so với hàng Normal.

- Đối với đơn Reseller thì phần đơn hàng được pickup khung 11H ngày N và create vào 20H (6,2%) cùng ngày hoặc vào 11H (6,1%), 14H (5,2%) ngày N+1. Đối với đơn ở khung 20H ngày N thì thường được tạo vào khung 20H (5,4%) ngày N+1 hoặc vào khung giờ 11H (6,4%) và 14H (4,8%) các ngày sau đó (N+2, N+3,...)

- Ngoài ra đối với đơn pickup vào 17H ngày N thì thường được create vào 11H (4,4% của N+1 và 4,1% cho allN) các ngày sau đó.

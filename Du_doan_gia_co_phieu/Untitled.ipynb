```python
# lấy trung bình và độ lệch của 5 ngày và 3 ngày như trên nhưng là volume
```

```python
Pre_volume_5_std = []
Pre_volume_5_avg = []

for index, row in df.iterrows():
    if index < 5:
        Pre_volume_5_std.append(0)
        Pre_volume_5_avg.append(0)
        
    else:
        # gia tri trung binh va lech của 5 ngày gần nhất
        i_5 = index - 5
            
#         giá trị lệch
        tmp_std = df.iloc[i_5:index]['Volume'].std()
        Pre_volume_5_std.append(tmp_std)  
        
#         giá trị trung bình
        tmp_avg = df.iloc[i_5:index]['Volume'].mean()
        Pre_volume_5_avg.append(tmp_avg) 
        
df['Pre_volume_5_std'] = Pre_volume_5_std 
df['Pre_volume_5_avg'] = Pre_volume_5_avg


Pre_volume_30_std = []
Pre_volume_30_avg = []

for index, row in df.iterrows():
    if index < 30:
        Pre_volume_30_std.append(0)
        Pre_volume_30_avg.append(0)
    else:
        # gia tri trung binh va lech của 5 ngày gần nhất
        i_30 = index - 30
            
#         giá trị lệch
        tmp_std = df.iloc[i_30:index]['Volume'].std()
        Pre_volume_30_std.append(tmp_std)  
        
#         giá trị trung bình
        tmp_avg = df.iloc[i_30:index]['Volume'].mean()
        Pre_volume_30_avg.append(tmp_avg) 
        
df['Pre_volume_30_std'] = Pre_volume_30_std
df['Pre_volume_30_avg'] = Pre_volume_30_avg
```

```python
# them mot feature ratio pre_close_5_avg/pre_close_30_avg ratio_pre_5_over_3_avg
```

```python
df['Pre_Close_30_avg'] = df['Pre_Close_30_avg'] + 0.001
```

```python
df['ratio_pre_5_over_3_avg'] = df['Pre_Close_5_avg'] / df['Pre_Close_30_avg']
```

```python
# BackTesting

# train, test
from datetime import datetime 

train = df[df.Date < datetime(year = 2013, month = 1, day = 1)]
test = df[df.Date >= datetime(year = 2013, month = 1, day = 1)]
```

```python
selected_cols = ['Pre_Close_5_avg', 'Pre_Close_5_std', 'Pre_Close_30_avg', 
       'Pre_Close_30_std', 'Pre_volume_5_std', 'Pre_volume_5_avg',
       'Pre_volume_30_std', 'Pre_volume_30_avg', 'ratio_pre_5_over_3_avg']
```

```python
df[selected_cols].info()
```

```python
# train[selected_cols]
```

```python
X = df[selected_cols].copy()
y = df['Close']
```

```python
X = X.fillna(0)
y = y.fillna(0)
```

```python
# training a linear regression model
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# -- X là những feature để huấn luyện --> train[selected_cols]
# -- Y là những label ứng với từng Row --> train['Close']

# fitting
model.fit(X, y)
```

```python
# mỗi mô hình đều có thông số (gọi là thông số tự do)
```

```python
## prediction
predictions = model.predict(test[selected_cols])
```

```python
# # predictions
from sklearn.metrics import mean_absolute_error
mea =  mean_absolute_error(test['Close'], predictions)

# # ground truth value -- giá trị thực
# # giá trị dự báo
# metric_xyz(X, Y)

predictions[:10]
```

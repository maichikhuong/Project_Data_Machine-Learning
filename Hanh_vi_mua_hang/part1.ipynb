```python
## khi học ra kết quả là 0 và 1
# ví dụ học ra 2 người cùng 1
# nguoi A -- 1
# nguoi B -- 1

# giả sử chỉ tiếp cận được một người 

# 0, 1 --> nguoi A --> 0.9
# --> nguoi B --> 0.75

# 0.4, 0.5, 0.7, 0.9 --> 1, 0 --> threshoding, 0.6 -- 1, 0.6 -->0

# 1, 0 ------> 1,0

# 1 1 0 0 ---> accuracy 100  75%

# test -- 0.8 --> threshoding --> 0.2 --> 1 positive
```
